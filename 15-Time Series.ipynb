# Data
## Project description
Sweet Lift Taxi company has collected historical data on taxi orders at airports. To attract more drivers during peak hours, we need to predict amount of taxi orders for next hour. Build a model for such a prediction.
RMSE metric on test set should not be more than 48.

## Data description
**Features**
- `datetime`â€” observation time stamp.

**Target**
- `num_orders`- number of taxi orders at particular time stamp.

## Library imports and project environment setup

## Preprocessing

```python
#Duplicates
df.index.duplicated().sum()
```

```python
#Dropping unnecessary columns we used for plot above
df = df.drop(['mean','std'], axis=1)
```

## New features

```python
#Creating new features to help model make better predictions 
def make_features(data, max_lag, rolling_mean_size):
    data['month'] = data.index.month
    data['day'] = data.index.day
    data['dayofweek'] = data.index.dayofweek
    data['hourofday'] = data.index.hour
    
    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)

    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()
    #data['rolling_std'] = data['num_orders'].shift().rolling(rolling_mean_size).std()
    
make_features(df, 10, 5)

df.head()
```

# Model
## Splitting

```python
df = df.dropna(how='any', axis=0)

X = df.drop('num_orders', axis=1)
y = df['num_orders']

X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=False, test_size=0.1)

print('Train set from', X_train.index.min(), 'until', X_train.index.max())
print('Test set from', X_test.index.min(), 'until', X_test.index.max())
```

## Standard scaling

```python
sc = ss()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

## Model selection by RMSE

```python
#Creating function and scorer for GridSearch and cross-validation
def rmse(actual, predict):
    predict = np.array(predict)
    actual = np.array(actual)

    distance = predict - actual

    square_distance = distance ** 2

    mean_square_distance = square_distance.mean()

    score = np.sqrt(mean_square_distance)

    return score

rmse_scorer = make_scorer(rmse, greater_is_better = False)
```

```python
#TimeSeriesSplit for models to split train set into train and validation sets
tscv = TimeSeriesSplit()
```

### Baseline

```python
model = DummyRegressor(strategy='median')
baseline = np.mean(abs(cross_val_score(model,  X_train, y_train, cv=tscv, scoring = rmse_scorer)))
baseline
```

### Random Forest
#### Base

```python
RF = RandomForestRegressor(random_state=12345)
RF_base = np.mean(abs(cross_val_score(RF,  X_train, y_train, cv=tscv, scoring = rmse_scorer)))
RF_base
```

#### Tuned

```python
params = {"n_estimators" : [500, 700],
             "max_depth" : [6, 7, 8, 9, 10]}

gsSVR = GridSearchCV(estimator=RF, cv=tscv, param_grid=params, n_jobs=-1, verbose=0, scoring=rmse_scorer)
gsSVR.fit(X_train, y_train)
SVR_best = gsSVR.best_estimator_

best_param = pd.DataFrame(gsSVR.best_params_, index=[0])
RF_score_tuned = abs(gsSVR.best_score_)
best_param['score'] = RF_score_tuned

best_param
```

### XGBoost
#### Base

```python
XGB = XGBRegressor(n_jobs=-1, random_state=12345)
XGB_base = np.mean(abs(cross_val_score(XGB,  X_train, y_train, cv=tscv, scoring = rmse_scorer)))
XGB_base
```

#### Tuned

```python
params = {"n_estimators" : [500, 700],
             "max_depth" : [6, 7, 8, 9, 10]}

gsSVR = GridSearchCV(estimator=XGB, cv=tscv, param_grid=params, n_jobs=-1, verbose=0, scoring=rmse_scorer)
gsSVR.fit(X_train, y_train)
SVR_best = gsSVR.best_estimator_

best_param = pd.DataFrame(gsSVR.best_params_, index=[0])
XGB_score_tuned = abs(gsSVR.best_score_)
best_param['score'] = XGB_score_tuned

best_param
```

### LightGBM
#### Base

```python
LGB = LGBMRegressor(random_state=12345)
LGB_base = np.mean(abs(cross_val_score(LGB,  X_train, y_train, cv=tscv, scoring = rmse_scorer)))
LGB_base
```

#### Tuned

```python
params = {"n_estimators" : [500, 700],
             "max_depth" : [6, 7, 8, 9, 10]}

gsSVR = GridSearchCV(estimator=LGB, cv=tscv, param_grid=params, n_jobs=-1, verbose=0, scoring=rmse_scorer)
gsSVR.fit(X_train, y_train)
SVR_best = gsSVR.best_estimator_

best_param = pd.DataFrame(gsSVR.best_params_, index=[0])
LGB_score_tuned = abs(gsSVR.best_score_)
best_param['score'] = LGB_score_tuned

best_param
```

### CatBoost
#### Base

```python
CB = CatBoostRegressor(verbose=0, loss_function="RMSE", random_state=12345)
CB_base = np.mean(abs(cross_val_score(CB,  X_train, y_train, cv=tscv, scoring = rmse_scorer)))
CB_base
```

#### Tuned

```python
params = {"iterations" : [500, 700],
             "depth" : [6, 7, 8, 9, 10]}

gsSVR = GridSearchCV(estimator=CB, cv=tscv, param_grid=params, n_jobs=-1, verbose=0, scoring=rmse_scorer)
gsSVR.fit(X_train, y_train)
SVR_best = gsSVR.best_estimator_

best_param = pd.DataFrame(gsSVR.best_params_, index=[0])
CB_score_tuned = abs(gsSVR.best_score_)
best_param['score'] = CB_score_tuned

best_param
```

### Results

```python
#Plotting model scores
models = pd.DataFrame({
    'Model': ['baseline', 'RF_base', 'XGB_base', 'LGB_base', 'CB_base', 
             'RF_tuned', 'XGB_tuned', 'LGB_tuned', 'CB_tuned'],
    'Score': [baseline, RF_base, XGB_base, LGB_base, CB_base,
             RF_score_tuned, XGB_score_tuned, LGB_score_tuned, CB_score_tuned]})
sorted_by_score = models.sort_values(by='Score', ascending=False)

sns.barplot(x='Score', y = 'Model', data = sorted_by_score, color = 'g')
plt.title('ML models')
plt.xlabel('RMSE- test data (%)')
plt.ylabel('Model');
```

Plot depicts LGB_base has best train RMSE (25.015).

## LGB- chosen model
### Features' importances

```python
#Finding features' importances for prediction
LGB = LGBMRegressor(random_state=12345)
LGB.fit(X_train, y_train)

LGB_feature_importance = pd.Series(LGB.feature_importances_, df.drop(['num_orders'], axis=1).columns).sort_values(ascending=True)
plt.barh(LGB_feature_importance.index, LGB_feature_importance)
plt.xlabel("LightGBM Feature Importance");
```

`hourofday` is most important feature.

### Testing

```python
y_pred_test = LGB.predict(X_test)
LGB_test_score = rmse(y_test, y_pred_test)
LGB_test_score
```

```python
round((LGB_test_score-LGB_base)/LGB_test_score * 100, 2)
```

Test score is 42.24% higher than train score, maybe due to slight overfitting. This leaves room for future improvement, yet target of 48 or lower has been met.

### Sanity check

```python
model = DummyRegressor(strategy='median')
model.fit(X_train, y_train)
y_base_test = model.predict(X_test)
base_test_score = rmse(y_test, y_base_test)
base_test_score
```

```python
round((base_test_score-LGB_test_score)/base_test_score * 100, 2)
```

Final CatBoost test score is 50.36% lower than dummy baseline, meaning modeling was useful.

```python
plt.figure(figsize=(10,3))
y_test.plot(label='y_test')
pd.Series(y_pred_test, index=y_test.index).plot(label='prediction')
plt.title('y_test vs prediction')
plt.legend()
plt.show()
```
